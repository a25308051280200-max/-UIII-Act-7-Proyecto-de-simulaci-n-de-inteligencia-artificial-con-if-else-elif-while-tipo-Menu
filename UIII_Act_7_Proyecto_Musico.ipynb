### Musico y Cancion

```python
# --- SISTEMA DE IA: LABORATORIO MUSICAL PARA ESTUDIANTES ---
print("BIENVENIDO AL ASISTENTE VIRTUAL PARA MÚSICOS v1.0")

# Variable para controlar el menú principal y mantener el programa activo [2]
opcion_menu = ""

# Bucle while para que el programa se repita hasta que el usuario decida salir [2, 3]
while opcion_menu != "6":
    print("\n" + "="*40)
    print("--- MENÚ DE INTELIGENCIA ARTIFICIAL ---")
    print("1.  Predicción de Éxito Comercial")
    print("2. Análisis de Estado de Ánimo (Mood)")
    print("3.  Evaluación de Complejidad Técnica")
    print("4.  Recomendación de Género Musical")
    print("5.  Prueba de Aptitud para Radio")
    print("6. Salir del Laboratorio")
    print("="*40)

    opcion_menu = input("Seleccione un módulo de IA (1-6): ")

    # Estructura selectiva para dirigir el flujo del programa [1, 4]
    if opcion_menu == "1":
        print("\n--- MÓDULO 1: PREDICCIÓN DE ÉXITO ---")
        nombre_cancion = input("Nombre de la canción: ")
        ritmo = int(input("Nivel de ritmo/pegajosidad (1-10): "))
        seguidores = int(input("Seguidores actuales del músico: "))

        # Uso de operadores aritméticos y lógicos para el veredicto [5-7]
        if ritmo >= 8 and seguidores > 1000:
            print(f"musico: '{nombre_cancion}' será un HIT VIRAL.")
        elif ritmo >= 5 or seguidores > 500:
            print(f"musico: '{nombre_cancion}' tendrá una audiencia MODERADA.")
        else:
            print("Resultado IA: Necesitas trabajar más en la promoción de la canción.")

    elif opcion_menu == "2":
        print("\n--- MÓDULO 2: ANÁLISIS DE ESTADO DE ÁNIMO ---")
        bpm = float(input("Velocidad de la canción (solo el número, ej. 95): ")) # Latidos por minuto
        intensidad = int(input("Nivel de energía (1-10): "))

        # La IA clasifica la canción según la velocidad y energía [4, 8]
        if bpm > 120 and intensidad >= 7:
            print("Esta es una canción ENÉRGICA y ALEGRE.")
        elif bpm < 90 and intensidad <= 4:
            print(" Esta es una canción RELAJANTE o TRISTE.")
        else:
            print(" Es una canción con un perfil NEUTRAL.")

    elif opcion_menu == "3":
        print("\n--- MÓDULO 3: COMPLEJIDAD TÉCNICA ---")
        experiencia = int(input("Años de experiencia del músico: "))
        acordes = int(input("¿Cuántos acordes distintos tiene la canción?: "))

        # Uso de comparaciones relacionales para evaluar la técnica [6]
        if acordes > 10 and experiencia > 5:
            print(" Composición de nivel MAESTRO / JAZZ.")
        elif acordes >= 4 and acordes <= 10:
            print("Composición de nivel ESTÁNDAR / POP.")
        else:
            print(" Es una pieza SIMPLE, ideal para principiantes.")

    elif opcion_menu == "4":
        print("\n--- MÓDULO 4: RECOMENDACIÓN DE GÉNERO ---")
        usa_distorsion = input("¿Utiliza guitarras con distorsión? (si/no): ").lower()
        usa_sintetizador = input("¿Utiliza sintetizadores electrónicos? (si/no): ").lower()

        # La IA deduce el género basándose en los instrumentos [1]
        if usa_distorsion == "si" and usa_sintetizador == "no":
            print(": El género ideal para esta canción es ROCK.")
        elif usa_distorsion == "no" and usa_sintetizador == "si":
            print(": El género ideal para esta canción es SYNTH-POP.")
        elif usa_distorsion == "si" and usa_sintetizador == "si":
            print( ":Tienes un estilo de FUSIÓN ALTERNATIVA.")
        else:
            print("Recomendación IA: Podría ser una pieza ACÚSTICA o FOLK.")

    elif opcion_menu == "5":
        print("\n--- MÓDULO 5: APTITUD PARA RADIO ---")
        duracion = float(input("Duración de la canción en minutos (ej. 3.5): "))

        # Las radios suelen preferir canciones de cierta duración [4]
        if duracion >= 2.5 and duracion <= 4.0:
            print("Resultado IA: CANCIÓN PERFECTA para reproducción en radio.")
        elif duracion < 2.5:
            print("Resultado IA: Demasiado corta; podría ser un JINGLE comercial.")
        else:
            print("Resultado IA: Demasiado larga; se recomienda una 'Radio Edit'.")

    elif opcion_menu == "6":
        print("\nFinalizando IA... ¡Gracias por crear música con nosotros!")

    else:
        # Manejo de opciones no válidas [1]
        print("\n[Error] Selección no válida. Por favor, intenta de nuevo.")

print("Programa rializado por Alejandra Rivera Vlaenzuela NL = 49.")
```
