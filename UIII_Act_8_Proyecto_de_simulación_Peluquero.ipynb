### Peluquero : peine

```python
# --- IA PELUQUERO VIRTUAL: EL ASISTENTE DEL PEINE ---
print("BIENVENIDO A LA BARBERÍA INTELIGENTE 2026")

# Variable para controlar el menú y mantener el ciclo activo
opcion_menu = ""

# Estructura repetitiva while para mantener el menú interactivo [4, 5]
while opcion_menu != "3":
    print("\n" + "="*40)
    print("--- MENÚ DEL PELUQUERO ---")
    print("1. Consultar recomendación de corte")
    print("2. Analizar estado del peine (IA Tool)")
    print("3. Salir de la peluquería")
    print("="*40)

    opcion_menu = input("Seleccione una opción (1-3): ")

    # Estructura selectiva simple y doble para manejar las opciones [6-8]
    if opcion_menu == "1":
        print("\n--- MÓDULO 1: RECOMENDACIÓN DE CORTE ---")

        # Entrada de datos interactiva [9]
        largo_pelo = float(input("¿Cuántos centímetros mide tu cabello actualmente?: "))
        estilo = input("¿Prefieres un estilo 'clasico' o 'moderno'?: ").lower()
        evento = input("¿Tienes algún evento importante hoy? (si/no): ").lower()

        # Lógica de la IA utilizando operadores relacionales y lógicos [10-13]
        if largo_pelo < 3 and estilo == "clasico":
            print("Resultado IA: Te recomiendo un 'Buzz Cut'. Es formal y siempre luce bien.")

        elif largo_pelo >= 3 and largo_pelo <= 10:
            # Uso de if anidado para mayor complejidad lógica [14, 15]
            if estilo == "moderno" or evento == "si":
                print("Resultado IA: El 'Pompadour' es ideal para tu largo y te hará destacar.")
            else:
                print("Resultado IA: Un 'Corte César' es práctico para tu día a día.")

        elif largo_pelo > 10 and estilo == "moderno":
            print("Resultado IA: Definitivamente un 'Man Bun' o melena con textura.")

        else:
            # Caso por defecto si no encaja en los perfiles específicos [16, 17]
            print("Resultado IA: Mantén el largo actual, pero pide un recorte de puntas.")

    elif opcion_menu == "2":
        print("\n--- MÓDULO 2: ANÁLISIS DEL PEINE ---")
        grosor_pelo = int(input("En una escala del 1 al 10, ¿qué tan grueso es tu cabello?: "))
        nudos = input("¿Tienes muchos nudos hoy? (si/no): ").lower()

        # La IA decide qué tipo de peine usar según los datos [16, 18]
        if grosor_pelo > 7 and nudos == "si":
            print("Acción IA: Utilizando PEINE DE PÚAS ANCHAS para evitar tirones.")
        elif grosor_pelo <= 7 and nudos == "no":
            print("Acción IA: Utilizando PEINE DE PÚAS FINAS para un acabado perfecto.")
        else:
            print("Acción IA: Utilizando PEINE MIXTO para equilibrar el volumen.")

    elif opcion_menu == "3":
        print("\n¡Gracias por visitarnos! El Peluquero IA te espera pronto.")

    else:
        # Manejo de opciones no válidas en el menú [19, 20]
        print("\n[Error] Opción no válida. Por favor, elige 1, 2 o 3.")
        print("Programa Realizado Por Alejandra Rivera")
```
